## BREAST CANCER DETECTION

### IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Data Collection and Processing

```python
#Loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#Loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data , columns = breast_cancer_dataset.feature_names)
```

```python
#Printing the first five rows
data_frame.head(10)
```

```python
print(data_frame.shape)     # number of rows, columns
print(data_frame.columns)   # list of columns
print(data_frame.empty)     # True if empty
```

```python
#Adding the target column to the dataset
data_frame['Label'] = breast_cancer_dataset.target
```

```python
#Print last 5 rows of the data frame
data_frame.tail()
```

```python
data_frame.shape
```

```python
# Getting some information about the data
data_frame.info()
```

```python
#statistical measures about the data
data_frame.describe()
```

```python
#checking the distribution of target variable
data_frame['Label'].value_counts()
```

##### 1 --> Benign
##### 0 --> Malignant

```python
data_frame.groupby('Label').mean()
```

### Seperating the features and target

```python
X = data_frame.drop(columns='Label', axis = 1)
Y = data_frame['Label'] 
```

```python
print(X)
```

```python
print(Y)
```

```python
Y.head()
```

### Splitting the data into training data & testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

### Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the logistic regression model using training data

model.fit(X_train,Y_train)
```

### Model Evaluation

##### Acurracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy on training data = ',training_data_accuracy*100,'%')
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
```

```python
print('Accuracy on testing data = ',test_data_accuracy*100,'%')
```

### Building a predictive system

```python
input_data =(20.29,14.34,135.1,1297,0.1003,0.1328,0.198,0.1043,0.1809,0.05883,0.7572,0.7813,5.438,94.44,0.01149,0.02461,0.05688,0.01885,0.01756,0.005115,22.54,16.67,152.2,1575,0.1374,0.205,0.4,0.1625,0.2364,0.07678
)

# change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one datapoint
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
    print('The Breast cancer is malignant.')

else:
    print('The Breast cancer is benign.')
```
